```python
import matplotlib.pyplot as plt
import numpy as np
from collections import OrderedDict
plt.close()
#plt.figure(dpi=300)
plt.style.use('ggplot')
plt.rcParams['axes.facecolor']='white'
plt.rcParams['grid.color']='gainsboro'
f=open("set_data.csv","r")
next(f)
next(f)
ratio={}
for l in f:
    row=l.split(",")
    if row[0]+" "+row[5] not in ratio:
        ratio[row[0]+" "+row[5]]=1
    else:
        ratio[row[0]+" "+row[5]]+=1
f.close()
print(ratio)
```

```python
f=open("set_data.csv","r")
next(f)
next(f)
services=OrderedDict()
for l in f:
    row=l.split(",")
    if str(row[0]) not in services :
        services[str(row[0])]=[1,[]]
        services[str(row[0])][1].append(float(row[5]))
    else:
        services[str(row[0])][0]+=1
        services[str(row[0])][1].append(float(row[5]))
#services=sorted(services.items(), key=operator.itemgetter(0))    
f.close()
print(services)
```
