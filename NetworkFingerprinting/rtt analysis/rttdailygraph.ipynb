```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import pandas as pd
from collections import OrderedDict

f=open("rtt_daily.csv","r")
next(f)
rtt=OrderedDict()
rtt_sd=OrderedDict()
for l in f:
    row=l.split(",")
    rtt[row[0]]=float(row[1].strip())
f.close()

print("IP Range 172.16.68.0-32\n")
print("Time(24hrs): Mean RTT (ms)")
for fname in rtt:
    print(fname.replace(".xml",""),": ",rtt[fname])
```

```python
f=open("rtt_sd.csv","r")
next(f)

rtt_sd=OrderedDict()
for l in f:
    row=l.split(",")
    rtt_sd[row[0]]=float(row[1].strip())
f.close()

print("IP Range 172.16.68.0-32\n")
print("Time(24hrs): SD RTT (ms)")
for fname in rtt_sd:
    print(fname.replace(".xml",""),": ",rtt_sd[fname])
```

```python
plt.close()
#plt.figure(dpi=300)
plt.figure(figsize = (15,5))
plt.style.use('ggplot')
plt.rcParams['axes.facecolor']='white'
plt.rcParams['grid.color']='gainsboro'

plt.plot([i for i in  range(len(rtt))],[rtt[i] for i in  rtt],label='Mean RTT')
plt.scatter([i for i in  range(len(rtt))],[rtt[i] for i in  rtt])

# plt.plot([i for i in  range(len(rtt_sd))],[rtt_sd[i] for i in  rtt_sd],label='StD RTT')
# plt.scatter([i for i in  range(len(rtt_sd))],[rtt_sd[i] for i in  rtt_sd])

# plt.legend(loc='upper right')
plt.xticks([i for i in  range(len(rtt))],[i.replace(".xml","") for i in  rtt])
plt.xlim(0,6)
plt.ylabel("Mean rtt")
plt.xlabel("Time (in 24 hrs)")
plt.title(" RTT Variation within 4-hours")
plt.show()
```
