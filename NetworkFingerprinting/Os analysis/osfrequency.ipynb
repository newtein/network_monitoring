```python
import matplotlib.pyplot as plt
f=open("set_dataos.csv","r")
next(f)
next(f)
os={}
for l in f:
    row=l.split(",")
    if row[1] not in os:
        os[row[1]]=1
    else:
        os[row[1]]+=1
f.close()
for items in os:
    print(items," : ",os[items])
```

```python
plt.clf()
plt.close()
plt.style.use('ggplot')
plt.rcParams['axes.facecolor']='white'
plt.rcParams['grid.color']='gainsboro'
plt.figure(figsize=(17,5))
plt.bar([i for i in range(len(os))],[os[i] for i in os])
plt.xticks([i for i in range(len(os))], [i for i in os],rotation=30)
plt.xlabel("Operating System")
plt.ylabel("Instances")
plt.title("Operating System Frequency")
plt.show()
```
